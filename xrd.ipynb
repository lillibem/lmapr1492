```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
# Télécharger le dossier TiSnPd.cif
with MPRester("pBDutDDdaCLFGZOVupy8") as m:

    structure = m.get_structure_by_material_id("mp-961682")

structure.to(filename="TiSnPd.cif")
```

```python
xrd_calculator = XRDCalculator(1.54060, 1)
pattern = xrd_calculator.get_pattern(structure)
```

## Trois premiers pics du diffractogramme pour un rayonnement X incident CuKa de longueur d’onde λ=1.54060Å 

```python
pics = ['Premier', 'Deuxième', 'Troisième']
for i in range(3) : 
    print("{} pic : \n Intensité : {} % \n Angle : 2\u0398 = {} ° \n Indices hkl associés : {} de multiplicité {} \n".format(pics[i], pattern.y[i], pattern.x[i], pattern.hkls[i][0]['hkl'], pattern.hkls[i][0]['multiplicity']))
```

```python
xrd_calculator.show_plot(structure, two_theta_range=(20,48))
```
